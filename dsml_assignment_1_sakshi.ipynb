```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
# reading the dataset
fires = pd.read_csv('fires.csv')
fires.head()
```

```python
# getting the dimensions of dataframe 
fires.shape
```

```python
# making the column names lowercase
fires.columns = fires.columns.str.lower()
fires.head()
```

```python
#capitalizing the first letter of data in fire_name column
fires['fire_name'] = fires['fire_name'].str.title()
fires.head()
```

```python
# renaming fires_size column
fires.rename(columns = {'fire_size' : 'acres_burned'}, inplace= 'true')
fires
```

```python
# droping rows having size less than 10 acres
fires.drop(fires[fires['acres_burned'] < 10].index, inplace=True)
fires
```

```python
# droping duplicate rows
fires = fires.drop_duplicates()
fires
```

```python
fires
```

```python
# changing the datatype of 'contain_date' and 'discovery_date' columns
fires['contain_date']= pd.to_datetime(fires['contain_date'])
fires['discovery_date']= pd.to_datetime(fires['discovery_date'])
fires.info()
fires

```
